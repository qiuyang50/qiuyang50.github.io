```python
import numpy                as np
import matplotlib.pyplot    as plt 
```

```python
xg   = np.linspace(0,1,400,endpoint=False)
fun1 = np.sin(2*np.pi*9*xg)
fun2 = np.sin(2*np.pi*xg)

xgcg   = np.linspace(0,1,8,endpoint=False)
fun1cg = np.sin(2*np.pi*9*xgcg)
fun2cg = np.sin(2*np.pi*xgcg)
```

```python
plt.rcParams['font.size'] = '30'
fig, ax = plt.subplots(1,1,figsize=(30,12)) 
plt.subplots_adjust(left=0.050,bottom=0.040,right=0.990,top=0.950,wspace=0.06,hspace=0.07)

ax.plot(xg,fun1,'-b',lw=2,label=r'$\sin(2\pi9x)$')
ax.plot(xgcg,fun1cg,'bo',mfc='none',ms=30,mew=3)
ax.plot(xg,fun2,'-r',lw=2,label=r'$\sin(2\pi x)$')
ax.plot(xgcg,fun2cg,'rs',mfc='none',ms=40,mew=3)
ax.legend(loc=1)
ax.grid()
ax.set_xlim([-0.05,1.05])
ax.set_ylim([-1.1,1.1])
ax.set_title('Example of Aliasing')

plt.savefig('aliasing.png')
```
